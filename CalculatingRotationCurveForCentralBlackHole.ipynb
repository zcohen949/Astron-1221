```python
import astropy.units as u
import astropy.constants as ac
import numpy as np
```

How to define a function in Python?

```python
def calculatingOrbitalVelocity(M, R):
  """
  This function calculates the orbital velocity
  Input:
  M, the mass of the central object
  R, the orbital radius of the satellite
  Output:
  orbitalVelocity, the orbital velocity of the satellite
  """
  orbitalVelocity = np.sqrt(ac.G * M / R)
  return orbitalVelocity
```

```python
# Below we define input M and R
M = 1 * u.earthMass
R = 1 * u.earthRad

# Calculating orbital velocity of an earth satellite
v_earthSatellite = calculatingOrbitalVelocity(M, R) # using the predefined function
print(v_earthSatellite.to(u.km/u.s))
```

We are going to introduce the concept of an array

```python
# Orbital radius of the eight solar system planets from inside out
# Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune
R_planet_arr = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6, 19.2, 30.1]) * u.au
```

```python
# Now we calculate the orbital velocities for eight planets
M_sun = 1 * u.solMass
v_planets_arr = calculatingOrbitalVelocity(M_sun, R_planet_arr)
print(v_planets_arr.to(u.km/u.s))
```

```python
R_planets_arr = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6, 19.2, 30.1]) * u.au
```

```python
# prompt: Can you help me with plotting v_planet_arr and R_planet_arr?

import matplotlib.pyplot as plt # importing plotting package

plt.plot(R_planet_arr, v_planets_arr.to(u.km/u.s)) # Plotting velocity vs. orbital radius
plt.xlabel('Orbital Radius (AU)') # Add x axis label
plt.ylabel('Orbital Velocity (km/s)') # Add y axis label
plt.show()
```

Calculating Rotation Curve For Central Black Hole

```python
print(np.arange(1,30)) # made a range of numbers from 1 to 30, which will be used to mark different points in the Milky Way
```

```python
R_MilkyWay_arr = np.array((np.arange(1,30))) * u.kiloparsec
# made an array from the range of numbers and transfered them to different radii in the Milky Way in the units of kiloparsec
```

```python
M_CentralBlackHole = (4.297 * 1000000) * u.solMass
v_MilkyWay_arr = calculatingOrbitalVelocity(M_CentralBlackHole, R_MilkyWay_arr)
print(v_MilkyWay_arr.to(u.kiloparsec/u.s))
# Now we calculate the orbital velocities at each point in the range
```

```python
import matplotlib.pyplot as plt # importing plotting package

plt.plot(R_MilkyWay_arr, v_MilkyWay_arr.to(u.km/u.s)) # Plotting velocity vs. orbital radius
plt.xlabel('Orbital Radius (AU)') # Add x axis label
plt.ylabel('Orbital Velocity (km/s)') # Add y axis label
plt.show() # makes the graph
```
