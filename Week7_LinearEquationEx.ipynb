```python
import numpy as np # imports the numpy package which allows us to conduct mathematical operations for the defined quantities
import matplotlib.pyplot as plt # imports a plotting package
```

```python
Income_arr = np.array([36700.,46400.,50200.,54200.,75100.,89200.]) # creates an array from the values listed which represents each step of male income from their educational status
Education_arr = np.arange(len(Income_arr)) # puts the values into a list in this order for our education vs. income plot below
plt.plot(Education_arr,Income_arr, "o-") # plots a graph below from our education and income arrays
plt.xlabel("Education Level") # labels the x-axis as education level
plt.ylabel("Income [$]") # labels the y-axis as income
plt.show()
```

```python
Education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
]) # creates an array from this list which represents each step of male income from their educational status
```

```python
plt.plot(Education_arr,Income_arr, "o-")
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.show()
# creates the same graph as above but with education levels instead of numbers on the x-axis
```

```python
z_linear = np.polyfit(Education_arr,Income_arr,1) # takes the data and creates a line from the average education vs. income values
print("Slope = ", z_linear[0], " in dollar per level") # prints the slope and units below
print("Y-intercept = ", z_linear[1], " in dollar at level 0, i.e., Less than high school completion")  # prints the y-intercept and units below
```

```python
O = Income_arr
O_err = np.array([3110, 1070, 2000, 4130, 4110, 8910]) # creates an array from error values taken from data of the confidence of our previous values of male income vs. education
C_linear = z_linear[0] * Education_arr + z_linear[1] # creates a line using out slope and y-intercept

Income_arr = np.array([36700.,46400.,50200.,54200.,75100.,89200.]) # creates the same array of values from before showing income per educational level
Education_arr = np.arange(len(Income_arr)) # makes this array two dimensional
plt.errorbar(Education_arr, O, yerr=O_err, marker="o", linestyle = "None", color = "black", label = "Data") # plots error bars at each education level from our education and 0 arrays
plt.plot(Education_arr,C_linear, color = "red", label = "Linear fit") # plots our line
plt.legend() # creates a legend
plt.xlabel("Education Level") # labels the x-axis
plt.ylabel("Income [$]") # labels the y-axis
plt.show()
```
